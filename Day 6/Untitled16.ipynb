```python
x=x/(np.max(x)) 
y=y/(np.max(y)) 
```

```python
m,c = np.random.random(), np.random.random()   #output variables, we have to tune m and c
```

```python
def learn (x,y,m,c,epoch):
    for i in range (epoch):
        error = y - (m*x + c)
        x_error = (y - (m*x + c))*x     #element wise multiplication will happen in this array
        error = np.sum(error)/len(x)     #error average
        x_error = np.sum(x_error)/len(x)
        learning_rate = 0.01
        delta_m = learning_rate*x_error #input times error 
        delta_c = learning_rate*error
        #to reflect these changes in m and c we have to add these changes to m and c
        m += delta_m
        c += delta_c
    return (m,c)
```

```python
m,c = learn(x,y,m,c,20000)
```

```python
x1 = [0,1]
y1 = [0*m+c , 1*m+c]
```
