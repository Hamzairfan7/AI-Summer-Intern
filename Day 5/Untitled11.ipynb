# REAL WORLD EXAMPLES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### GAS PRICES

```python
gas = pd.read_csv(r'F:\internship\Day 5\gas_prices.csv')
```

```python
plt.title('Gas prices in years')
plt.plot(gas.Year, gas['USA'], 'b.-', label = 'USA')
plt.plot(gas.Year, gas['Canada'], label = 'Canada')
plt.plot(gas.Year, gas['Germany'], label = 'Germany')
plt.legend()
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.xlabel('Years')
plt.ylabel('US Dollors')
plt.show()
```

```python


for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'], gas[country], marker='.')

plt.xlabel('Year')
plt.ylabel('Gas Price')
plt.title('Gas Prices Over Years')
plt.xticks(gas.Year[::3])
plt.legend(gas.columns[1:]) 
plt.show()

    
```

# FIFA DATA SET

```python
fifa=pd.read_csv(r'F:\internship\Day 5\fifa_data.csv')
```

### HISTOGRAM

```python
bins =[40,50,60,70,80,90]
plt.hist(fifa.Overall, bins=bins, color= 'purple')
plt.xticks(bins)
plt.ylabel('num of players')
plt.xlabel('years')
plt.title('players skills score in FIFA 2018')

plt.show()
```

### PIE CHART

```python
left=fifa.loc[fifa['Preferred Foot']== 'Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]
```

```python
left
```

```python
labels = ['left','right']
colors=['r','g']
plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f')
plt.title('Foot Preferrence')
plt.show()
```
